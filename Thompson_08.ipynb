```python
import pandas as pd
import seaborn as sns
```

```python
avocado = pd.read_csv("avocado.csv")
avocado.head()
```

```python
avocado.info()
```

```python
avocado['Date'] = pd.to_datetime(avocado['Date'])
print(avocado.info())
```

```python
filtered_data = avocado.query('year == 2015 and type == "conventional" and region == "Albany"')
```

```python
filtered_data = filtered_data[['Date', 'Total Bags', 'Small Bags']]
```

```python
filtered_data.head()
```

```python
melted_avocado = filtered_data.melt(id_vars=['Date'], value_vars=['Total Bags', 'Small Bags'], var_name='Bags', value_name='Count')

# Display the melted data
print(melted_avocado)
```

```python
sns.lineplot(data=melted_avocado, x='Date', y='Count', hue='Bags')
```

```python
avocado['Sales Volume'] = pd.qcut(avocado['Total Volume'], q=4, labels=['Poor', 'Modest', 'Good', 'Excellent'])
avocado.head
```

```python
sns.countplot(data=avocado, x='year', hue='Sales Volume')
```

```python
melted_avocado = pd.melt(avocado, id_vars=['year'], value_vars=['Small Bags', 'Large Bags', 'XLarge Bags'], var_name='Bags', value_name='Count')
melted_avocado
```

```python
sns.relplot(data=melted_avocado, kind='line', x='year', y='Count', hue='Bags')
```
